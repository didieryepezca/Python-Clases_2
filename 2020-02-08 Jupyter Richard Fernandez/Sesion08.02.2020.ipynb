# python visual analytics

```python
import os as os
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print("Directorio actual de trabajo: ", os.getcwd())
```

```python
df = pd.read_csv("mtcars.csv", sep = ",")
df.head()
```

```python
df.hist()
```

```python
plt.show()
```

```python
df.plot(kind='density',subplots = True, layout = (3,3), sharex = False)
plt.show()
```

```python
df.plot(kind='box',subplots = True, layout = (3,3), sharex = False)
plt.show()
```

```python
plt.scatter(df['mpg'], df['wt'])
```

```python
np.corrcoef(df['mpg'],df['disp'])
```

```python
corrmat = df.corr()
cg = sns.clustermap(corrmat, cmap = "YlGnBu", linewidths = 0.1);
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation = 0)
cg
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```

```python
corr = df.corr()
ax = sns.heatmap(
corr, vmin = 1, vmax = 1, center = 0,
cmap = sns.diverging_palette(20,220,n=200),
    square = True
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right'
);
```
