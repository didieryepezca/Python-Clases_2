# Filter selection

## t student

```python
from time import time
time_star = time()

list_t = []

from scipy import stats
features_t_num = []
for feature in features_num:
    t, p = stats.ttest_ind(train3[train3['FLG_DEAC_M2']==1][feature],train3[train3['FLG_DEAC_M2']==0][feature])
    list_t.append([feature,t,p])

df_t = pd.DataFrame(list_t, columns=['feature','t','p_valor'])

time_end = time()
print ("Time: ", np.round((time_end-time_star)/60,2), " minutes")
```

```python
lista_t = list(df_t[df_t['p_valor']<0.01].sort_values('p_valor', ascending = True)['feature'])
lista_t
```

## Chi cuadrado

```python
from time import time
time_star = time()

list_chi2 = []

from scipy import stats
features_tstudent = []
for feature in features_cat_dummies:
    X = train3[feature].astype(str)
    Y = train3['FLG_DEAC_M2'].astype(str)
    dfObserved = pd.crosstab(Y,X) 
    chi2, p, dof, expected = stats.chi2_contingency(dfObserved.values)
    list_chi2.append([feature,chi2,p])

df_chi2 = pd.DataFrame(list_chi2, columns=['feature','chi2','p_valor'])

time_end = time()
print ("Time: ", np.round((time_end-time_star)/60,2), " minutes")
```
